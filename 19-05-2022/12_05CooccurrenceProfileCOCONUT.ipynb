```python
df_split[0] = df_split[0] [['Bitset']]
```

```python
df_split[0]
```

```python
index = df_split[0].index
index
```

```python
for x in index:
    a = df_split[0]['Bitset'][x]
    a = a.split(", ")
    a = re.sub('\[', '', str(a))
    a = re.sub('\]', '', str(a))
    df_split[0]['Bitset'][x] = a
```

```python
feature_list0 = []
for x in df_split[0]['Bitset']:
    tup = literal_eval(x)
    feature_list0.append(tup)
feature_list0[-1]
```
