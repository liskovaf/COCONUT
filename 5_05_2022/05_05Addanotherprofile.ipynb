### Creating histogram

```python
#Loading zinc database from csv 
suppl_csv = pd.read_csv('../data/ZINCECFP.csv', delimiter = ' ')
suppl_csv
```

```python
suppl_csv  = suppl_csv [['Bitset']]
suppl_csv 
```

```python
for x in suppl_csv.index:
    a = suppl_csv['Bitset'][x]
    a = a.split(", ")
    a = re.sub('\[', '', str(a))
    a = re.sub('\]', '', str(a))
    suppl_csv['Bitset'][x] = a
```

```python
feature_list = []
for x in suppl_csv['Bitset']:
    tup = literal_eval(x)
    feature_list.append(tup)
feature_list[:3]
```

```python
feature_list[-1]
```

```python
p = CooccurrenceProfile.from_feature_lists(feature_list)
p.df
```

```python
p.df.to_csv('../data/CooccurenceProfile/ZINCp.csv')
pd.read_csv('../data/CooccurenceProfile/ZINCp.csv',index_col=[0,1])
```

```python
p = profiles.CooccurrenceProfile.from_dataframe(pd.read_csv('../data/CooccurenceProfile/ZINCp.csv'))
p.df
```

```python
print(f"There are {len(p.distinct_features())} distinct.")
```

```python
self_relations_values = p.select_self_relations()['value']
self_relations_values
```

```python
title = f"Stand-alone ECFPlike occurrence in ZINC"
a4_dims = (11.7, 8.27)
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=a4_dims)
sns.histplot(self_relations_values, log_scale=True)
ax.set(title=title, xlabel=f"Stand-alone occurrence count within ZINC", ylabel=f" ECFPlike")
plt.savefig(f"ZINC.svg", format="svg")
#plt.close()
```
